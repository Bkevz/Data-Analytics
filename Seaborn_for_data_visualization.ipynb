Data loading

```python
import seaborn as sns
import pandas as pd
dataset = pd.read_csv('data.csv')
```

1. Statistical Relationship - scatterplot

```python
sns.relplot(x='hours', y='marks', hue='age', data=dataset)
```

1.2 Statistical Relationship - Lineplot

```python
sns.relplot(x='hours', y='marks', hue='age', data=dataset, kind= "line", style="internet")
```

Statistical relationship for multiple data.

```python
sns.relplot(x='hours', y='marks', hue='age', data=dataset, col='internet') #separating data into two with parameter internet on dataset
```

2. Data Distribution by Histogram

```python
sns.displot(dataset, x='hours', binwidth=0.5, hue="internet")
```

Data distribution-Bivariate

```python
sns.displot(dataset, x="age", y="marks",hue="internet")
```

Kernel Density Function

```python
sns.displot(dataset, x="age", y="marks",kind="kde",hue="internet")
```

Pairplot.

```python
sns.pairplot(dataset) #Showing all the plots
```

Categorical Data- Scatterplot

```python
sns.catplot(x="age", y="marks", data=dataset, hue="internet")
```

```python
sns.catplot(x="age", y="marks", kind="box", data=dataset, hue="internet")
```

Regression for categorical data

```python
sns.regplot(x="marks", y="hours", data=dataset);

sns.lmplot(x="marks", y="hours", data=dataset,hue="internet",markers=["o", "x"], palette="Set1");
```

Using SNS theme in other graph

```python
import matplotlib.pyplot as plt
import numpy as np
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1, 10):
        plt.plot(x, np.sin(x + i  * .5) * (7 - i) * flip)
sns.set_context("talk") #talk/poster/paper
sinplot()
```
